```python
from tkinter import *

def itemsCost():
    mealCost = int(fries.get())*30+int(noodles.get())*30+int(soup.get())*10+int(burger.get())*25+int(sandwich.get())*35+int(drinks.get())*15
    myTextMeal.set(float(mealCost))
    
    serviceCharge = (mealCost * 15)/100
    myTextService.set(serviceCharge)
    
    salesTax = (mealCost * 5)/100
    myTextSales.set(salesTax)
    
    subTotal = serviceCharge + salesTax
    myTextSub.set(subTotal)
    
    total = mealCost + subTotal
    myTextTotal.set(total)
    
def reset():
    myTextMeal.set("")
    myTextService.set("")
    myTextSales.set("")
    myTextSub.set("")
    myTextTotal.set("")
    
    fries.delete(0,END)
    noodles.delete(0,END)
    soup.delete(0,END)
    burger.delete(0,END)
    sandwich.delete(0,END)
    drinks.delete(0,END) 
    
def closeWin():
    master.destroy()

master=Tk()

title = Label(master,text="Flavours Restaurent").grid(row=0,column=3,sticky=E)

myTextMeal=StringVar()
myTextService=StringVar()
myTextSales=StringVar()
myTextSub=StringVar()
myTextTotal=StringVar()

Label(master,text="Fries").grid(row=1,sticky=W)
Label(master,text="Noodles").grid(row=2,sticky=W)
Label(master,text="Soup").grid(row=3,sticky=W)
Label(master,text="Burger").grid(row=4,sticky=W)
Label(master,text="Sandwich").grid(row=5,sticky=W)
Label(master,text="Drinks").grid(row=6,sticky=W)

fries=Entry(master)
noodles=Entry(master)
soup=Entry(master)
burger=Entry(master)
sandwich=Entry(master)
drinks=Entry(master)

fries.grid(row=1,column=1)
noodles.grid(row=2,column=1)
soup.grid(row=3,column=1)
burger.grid(row=4,column=1)
sandwich.grid(row=5,column=1)
drinks.grid(row=6,column=1)

Label(master,text="30/-").grid(row=1,column=2)
Label(master,text="30/-").grid(row=2,column=2)
Label(master,text="10/-").grid(row=3,column=2)
Label(master,text="25/-").grid(row=4,column=2)
Label(master,text="35/-").grid(row=5,column=2)
Label(master,text="15/-").grid(row=6,column=2)

Label(master,text="Cost of meal").grid(row=1,column=4)
Label(master,text="Service Charge").grid(row=2,column=4)
Label(master,text="Sales Tax").grid(row=3,column=4)
Label(master,text="Sub Total").grid(row=4,column=4)
Label(master,text="Total Cost").grid(row=5,column=4)

meal=Entry(master,text="",textvariable=myTextMeal).grid(row=1,column=5,sticky=E)
service=Entry(master,text="",textvariable=myTextService).grid(row=2,column=5,sticky=E)
sales=Entry(master,text="",textvariable=myTextSales).grid(row=3,column=5,sticky=E)
sub=Entry(master,text="",textvariable=myTextSub).grid(row=4,column=5,sticky=E)
total=Entry(master,text="",textvariable=myTextTotal).grid(row=5,column=5,sticky=E)

Label(master,text="15%").grid(row=2,column=6)
Label(master,text="5%").grid(row=3,column=6)

b=Button(master,text="Total",command=itemsCost)
b.grid(row=8,column=1,columnspan=50,rowspan=40,sticky=W,padx=5,pady=10)

r=Button(master,text="Reset",command=reset)
r.grid(row=8,column=3,columnspan=50,rowspan=40,sticky=W,padx=5,pady=10)

e=Button(master,text="Exit",command=closeWin)
e.grid(row=8,column=5,columnspan=50,rowspan=40,sticky=S,padx=5,pady=10)

master.mainloop()
```
